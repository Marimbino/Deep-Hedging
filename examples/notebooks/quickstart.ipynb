# deephedge quickstart

Hedge a 30-day at-the-money European call on a GBM stock with an MLP hedger, then price it via utility indifference.

```python
from deephedge import BrownianStock, EuropeanOption, Hedger, MLPHedger

stock = BrownianStock(sigma=0.2, cost=1e-4)
option = EuropeanOption(stock, strike=1.0, maturity=30 / 252)
hedger = Hedger(MLPHedger(), option)
```

```python
history = hedger.fit(n_paths=20_000, n_epochs=100, batch_size=1024, fee=1e-4, seed=42)
history.tail()
```

```python
from deephedge.utils.plotting import plot_training_history

plot_training_history(history);
```

```python
price = hedger.price(n_paths=50_000, seed=0)
metrics = hedger.evaluate(n_paths=10_000, fee=1e-4, seed=1)
print(f"utility-indifference price: {price:.5f}")
metrics
```

Compare with the analytic Black-Scholes benchmark (driftless GBM is risk-neutral with $r = 0$):

```python
from deephedge import BlackScholesHedger

bs = BlackScholesHedger(r=0.0, sigma=0.2)
print(f"Black-Scholes price: {float(bs.price(1.0, 1.0, 30 / 252)):.5f}")
```
