# WGAN evaluation

Train the TCN WGAN-GP on S&P 500 log-returns and inspect the stylized facts of the generated paths.

```python
from deephedge import QuantWGANGenerator

gen = QuantWGANGenerator(seq_len=30, noise_dim=5)
history = gen.fit(
    ticker="^GSPC",
    start="2006-01-01",
    end="2025-12-31",
    epochs=200,
    batch=256,
    seed=42,
    verbose=True,
)
```

## Training diagnostics

Wasserstein distance, critic/generator losses, and the gradient penalty per epoch.

```python
from deephedge.augmentation import WGANTrainer

WGANTrainer.plot_history(history);
```

## Stylized facts

Marginal distribution, QQ-plot, autocorrelation of returns and squared returns, and the leverage effect — generated vs real.

```python
gen.evaluate(n_gen=2000);
```

## Synthetic price paths

```python
from deephedge.utils.plotting import plot_price_paths

prices = gen.generate(50_000)
print(prices.shape)
plot_price_paths(prices, n_plot=100);
```

```python
gen.save("./wgan_model")
```
